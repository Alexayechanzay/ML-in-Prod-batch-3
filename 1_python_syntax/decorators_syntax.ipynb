```python
import time
```

### Decorator is a function which takes base_function and returns an enhanced function

```python
# decorator function
def timer_dec(base_func):
    
     # Func which will enhance the base one.
    def enhanced_func():
        start_time = time.time()
        base_func()
        end_time = time.time()
        print(f"Task time: {end_time - start_time:.2f} seconds")
        
    return enhanced_func
```

```python
def brew_tea(): # Act as base_func
    print("Brewing tea...")
    time.sleep(1)
    print("Tea is ready!")
```

### 2 Ways to apply decorator to a particular function

```python
dec_brew_tea = timer_dec(brew_tea)
dec_brew_tea()
```

```python
@timer_dec
def brew_tea(): # Act as base_func
    print("Brewing tea...")
    time.sleep(1)
    print("Tea is ready!")
```

```python
@timer_dec
def make_matcha():
    print("Making matcha...")
    time.sleep(1)
    print("Matcha is ready!")
```

```python
make_matcha()
print("-----------------------")
brew_tea()
```

### Decorator with arguments

```python
# decorator function
def timer_dec(base_func):
    
     # Func which will enhance the base one.
    def enhanced_func(*args, **kwargs): # taking positional arbitary argument
        start_time = time.time()
        base_func(*args, **kwargs) # unpack *arg tuple into individual args
        end_time = time.time()
        print(f"Task time: {end_time - start_time:.2f} seconds")
        
    return enhanced_func
```

```python
@timer_dec
def brew_tea(tea_type, steep_time): # Act as base_func
    print(f"Brewing {tea_type} tea...")
    time.sleep(steep_time)
    print("Tea is ready!")
```

```python
@timer_dec
def make_matcha():
    print("Making matcha...")
    time.sleep(1)
    print("Matcha is ready!")
```

```python
brew_tea("Yellow", 3)
```

```python
make_matcha()
```

```python
brew_tea(tea_type="Green", steep_time=1)
```

## Problem 1

```python
def mission_timer(base_func):

    def wrapper():

        start_time = time.time()
        
        i = 3
        while i > 0: # 3-2-1
            print(i)
            time.sleep(1)
            i -= 1
            
        base_func()
        end_time = time.time()
        print(f"Missin duration: {end_time - start_time:.6f} seconds")

    return wrapper
```

```python
@mission_timer
def launch_probe():
    print("Launching probe...")
    time.sleep(1)
```

```python
@mission_timer
def deploy_satellite():
    print("Deploying satellite...")
    time.sleep(1)
```

```python
launch_probe()
print("--------------")
deploy_satellite()
```

```python
@mission_timer
def launch_probe(target):
    print(f"Launching probe toward {target}")
    time.sleep(1)
```

## Problem 3 (Handling return values)

```python
def mission_timer(base_func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        i = 3
        while i > 0: # 3-2-1
            print(i)
            time.sleep(1)
            i -= 1 
        result = base_func(*args, **kwargs)
        end_time = time.time()
        print(f"Missin duration: {end_time - start_time:.6f} seconds")
        return result
    return wrapper
```

```python
@mission_timer
def launch_probe(target):
    print(f"Launching probe toward {target}...")
    time.sleep(1)
    return f"Probe successfully en route to {target}"
```

```python
@mission_timer
def deploy_satellite(orbit_type, altitude_km):
    print(f"Deploying satellite into {orbit_type} orbit at {altitude_km} km...")
    time.sleep(2)
```

```python
deploy_satellite("Polar", 800)
```

```python
result = launch_probe("Europa")
print("Stored mission result:",result)
```
